# Basic Plotting with Matplotlib
### 1. Creating Simple Line Plots
The simplest plot in Matplotlib is the line plot. You can create one using `plt.plot()` to visualize the relationship between two sequences of data.

```python
import matplotlib.pyplot as plt

# Simple data to plot
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Creating a basic line plot
plt.plot(x, y)
plt.title("Simple Line Plot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.show()
```

In this example:
- `plt.plot()` is used to plot `y` against `x`.
- `plt.show()` displays the plot in the notebook.

### 2. Customizing Line Styles, Colors, and Markers
Matplotlib allows you to customize your plots with various line styles, colors, and markers. Here are a few examples:


```python
# Data
x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
y2 = [2, 4, 6, 8, 10]

# Customizing line style, color, and markers
plt.plot(x, y1, color='blue', linestyle='--', marker='o', label='Dashed Line with Circle Marker')
plt.plot(x, y2, color='red', linestyle='-', marker='s', label='Solid Line with Square Marker')

# Adding title and labels
plt.title("Customized Line Styles and Markers")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")

# Displaying legend
plt.legend()

# Show plot
plt.show()
```

- **Colors**: Use color names (`'blue'`), hexadecimal values (`'#FF5733'`), or abbreviations (`'r'` for red).
- **Line Styles**: Common styles include `'-'` (solid), `'--'` (dashed), `'-.'` (dash-dot), and `':'` (dotted).
- **Markers**: Represent data points with symbols like `'o'` (circle), `'s'` (square), `'^'` (triangle).

### 3. Plotting Multiple Lines
You can plot multiple lines on the same graph by calling `plt.plot()` multiple times before `plt.show()`.


```python
# Data for multiple lines
x = [1, 2, 3, 4, 5]
y1 = [1, 2, 4, 8, 16]
y2 = [1, 3, 6, 10, 15]
y3 = [2, 4, 6, 8, 10]

# Plotting multiple lines
plt.plot(x, y1, label='y = x^2', color='blue')
plt.plot(x, y2, label='y = x(x+1)/2', color='green')
plt.plot(x, y3, label='y = 2x', color='orange')

# Adding title and labels
plt.title("Multiple Lines on the Same Plot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")

# Displaying legend
plt.legend()

# Show plot
plt.show()
```

Using `plt.plot()` multiple times before `plt.show()` allows plotting multiple lines on the same plot. The `label` parameter provides an identifier for each line in the legend.

### 4. Adding Labels and Titles
A well-labeled plot makes it easier to understand. Use `plt.title()`, `plt.xlabel()`, and `plt.ylabel()` to add these elements to your plot.


```python
# Data
x = [0, 1, 2, 3, 4]
y = [0, 1, 4, 9, 16]

# Creating a plot with labels and title
plt.plot(x, y, marker='o', linestyle='-', color='purple')

# Adding title and axis labels
plt.title("Quadratic Growth")
plt.xlabel("Input Value (X)")
plt.ylabel("Output Value (Y)")

# Display the plot
plt.show()
```

- `plt.title()`: Adds a title to the plot.
- `plt.xlabel()`: Adds a label to the x-axis.
- `plt.ylabel()`: Adds a label to the y-axis.

### 5. Saving Plots to Files (PNG, PDF, SVG, etc.)
Matplotlib allows you to save plots in different formats using `plt.savefig()`.

```python
# Data for the plot
x = [0, 1, 2, 3, 4]
y = [0, 1, 4, 9, 16]

# Create the plot
plt.plot(x, y, marker='o', linestyle='-', color='green')

# Adding title and labels
plt.title("Saving a Plot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")

# Save the plot as a PNG file
plt.savefig("data/local/simple_plot.png")

# Save the plot as a PDF file
plt.savefig("data/local/simple_plot.pdf")

# Show the plot
plt.show()
```

- `plt.savefig("filename.extension")`: Saves the plot to the specified file format (e.g., PNG, PDF, SVG).
- You can use additional parameters like `dpi` for resolution and `bbox_inches='tight'` to remove excess whitespace.
