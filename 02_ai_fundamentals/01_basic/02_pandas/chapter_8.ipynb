## Data Visualization with Pandas

#### 1. Introduction to Data Visualization
- Pandas provides built-in capabilities for basic data visualization, primarily built on top of **Matplotlib**.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### 2. Line Plots
Line plots are useful for visualizing data over time or continuous trends.

```python
# Sample DataFrame for line plot
data = pd.DataFrame({
    'Date': pd.date_range(start='2023-01-01', periods=10),
    'Value': [1, 4, 6, 8, 12, 15, 18, 20, 22, 25]
})

# Creating a line plot
data.plot(x='Date', y='Value', kind='line')
plt.show()
```

### 3. Bar Plots
Bar plots are useful for categorical data comparison.

```python
# Sample DataFrame for bar plot
data_bar = pd.DataFrame({
    'Category': ['A', 'B', 'C', 'D'],
    'Values': [3, 7, 5, 9]
})

# Creating a bar plot
data_bar.plot(x='Category', y='Values', kind='bar')
plt.show()
```

### 4. Histograms
Histograms display the distribution of a single variable.

```python
# Generating random data for histogram
data_hist = pd.Series([1, 2, 2, 3, 3, 3, 4, 4, 5, 5, 5, 5])

# Creating a histogram
data_hist.plot(kind='hist', bins=5)
plt.show()
```

### 5. Scatter Plots
Scatter plots show the relationship between two numerical variables.

```python
# Sample DataFrame for scatter plot
data_scatter = pd.DataFrame({
    'X': [1, 2, 3, 4, 5],
    'Y': [2, 3, 4, 6, 8]
})

# Creating a scatter plot
data_scatter.plot(x='X', y='Y', kind='scatter')
plt.show()
```

### 6. Box Plots
Box plots are used for visualizing the distribution of data and detecting outliers.

```python
# Sample DataFrame for box plot
data_box = pd.DataFrame({
    'Category1': [1, 2, 3, 4, 5],
    'Category2': [3, 4, 5, 6, 7]
})

# Creating a box plot
data_box.plot(kind='box')
plt.show()
```

### 7. Customizing Plots (Title, Axis Labels, Style)
You can customize plots by adding titles, axis labels, legends, and changing styles.

```python
# Customizing a line plot
ax = data.plot(x='Date', y='Value', kind='line')
ax.set_title('Line Plot Example')
ax.set_xlabel('Date')
ax.set_ylabel('Values')
plt.show()
```

### 8. Integrating with Matplotlib and Seaborn
You can further customize Pandas plots by using Matplotlib or Seaborn for enhanced styling and visualization.

```python
# Using Matplotlib to customize Pandas plots
fig, ax = plt.subplots()
data.plot(x='Date', y='Value', kind='line', ax=ax, color='red')
ax.set_title('Customized Line Plot')
plt.show()
```

```python
import seaborn as sns

# Sample DataFrame for Seaborn integration
data_seaborn = pd.DataFrame({
    'Category': ['A', 'B', 'C', 'D'],
    'Values': [3, 7, 5, 9]
})

# Creating a Seaborn bar plot
sns.barplot(x='Category', y='Values', data=data_seaborn)
plt.show()
```

### 9. Plotting with Pandas Built-in Functions (plot(), hist(), etc.)
Pandas provides a variety of built-in functions to plot data quickly.

```python
data.plot(x='Date', y='Value', kind='line')
plt.show()
```

```python
data_hist.plot.hist(bins=5)
plt.show()
```

```python
data_bar.plot.bar(x='Category', y='Values')
plt.show()
```

```python
data_box.plot.box()
plt.show()
```
